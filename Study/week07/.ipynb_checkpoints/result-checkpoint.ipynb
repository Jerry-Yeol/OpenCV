```python
import cv2, sys
sys.path.append('..')
import numpy as np
from utils import *
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../lena.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img, (128,128), interpolation = cv2.INTER_CUBIC)
img2 = cv2.imread('../test.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img2[img2>0] = 1
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

# 이진화

```python
new_img = binarize(img, 128)
plt.imshow(new_img, cmap='gray')
plt.show()
```

# 오츄 알고리즘

```python
out, t = ostu(img)
plt.imshow(out, cmap='gray')
plt.show()
```

# 이진 모폴로지

## Erosion

```python
struct = np.array([[0,1,0],[0,1,0],[0,1,0]])
out = bin_erosion(img2,struct)
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.imshow(img2, cmap='gray')
plt.subplot(132)
plt.imshow(out, cmap='gray')
plt.subplot(133)
plt.imshow(img2-out, cmap='gray')
plt.show()
```

## Dilation

```python
out = bin_dilation(img2, struct)
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.imshow(img2, cmap='gray')
plt.subplot(132)
plt.imshow(out, cmap='gray')
plt.subplot(133)
plt.imshow(img2-out, cmap='gray')
plt.show()
```

## Open

```python
out = bin_opening(img2, struct)
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.imshow(img2, cmap='gray')
plt.subplot(132)
plt.imshow(out, cmap='gray')
plt.subplot(133)
plt.imshow(img2-out, cmap='gray')
plt.show()
```

## Close

```python
out = bin_closing(img2, struct)
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.imshow(img2, cmap='gray')
plt.subplot(132)
plt.imshow(out, cmap='gray')
plt.subplot(133)
plt.imshow(img2-out, cmap='gray')
plt.show()
```
