# Histogram

```python
hist,_ = to_hist(a)
plt.plot(hist,'b-')
plt.show()
```

# Histogram Equalization

```python
def pretty_hist(histogram):
    out = histogram.copy()
    for i, val in enumerate(histogram):
        if i==0 or i==(len(histogram)-1): continue
        if val == 0:
            out[i] = round((histogram[i-1]+histogram[i+1])/2)
            
    return out
            
    
```

```python
zxc = hist_eq(a)
plt.plot(pretty_hist(zxc[1]),'b-')
plt.show()
```

# Result

```python
plt.plot(hist, 'r-')
plt.plot(pretty_hist(zxc[1]), 'b-')

plt.show()
```

# Smoothing

## Box filter

```python
bfil = box_fil(5)
zxc=convolution(a, bfil)

plt.figure(figsize=(8,16))
plt.subplot(121)
plt.imshow(a, cmap='gray')
plt.subplot(122)
plt.imshow(zxc, cmap='gray')
plt.show()
```

## Gaussian filter

```python
gfil = gaussian_filter(5, 1)
zxc = convolution(a, gfil)

plt.figure(figsize=(8,16))
plt.subplot(121)
plt.imshow(a, cmap='gray')
plt.subplot(122)
plt.imshow(zxc, cmap='gray')
plt.show()
```

## Median filter

```python
zxc = median(a, (5,5))

plt.figure(figsize=(8,16))
plt.subplot(121)
plt.imshow(a, cmap='gray')
plt.subplot(122)
plt.imshow(zxc, cmap='gray')
plt.show()
```

# Sharpning

## Unsharp

```python
zxc = usNhb(a, 5, 1, 1)

plt.figure(figsize=(8,16))
plt.subplot(121)
plt.imshow(a, cmap='gray')
plt.subplot(122)
plt.imshow(zxc, cmap='gray')
plt.show()
```

## Laplacian

```python
lfil = laplacian(c=1, diagonal=False)
zxc = convolution(a, lfil)

plt.figure(figsize=(8,16))
plt.subplot(121)
plt.imshow(a, cmap='gray')
plt.subplot(122)
plt.imshow(np.double(a)-zxc, cmap='gray')
plt.show()
```
