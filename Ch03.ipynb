## Spatial operation

- Mask(Window)를 만들어 수행.
- Correlation : Mask 와 영상이 얼마나 비슷한지 측정.
- Convolution : Correlation 과 비슷하나 Mask 을 원점 대칭으로 뒤집어서 수행.

```python
fil = np.ones([5,5])/25
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.imshow(convolution(img, fil), cmap='gray')
plt.show()
```

```python
fil = np.ones([5,5])/25
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.imshow(correlation(img, fil), cmap='gray')
plt.show()
```
