```python
def gaussian_conv(img):
    
    h, w = img.shape
    
    fil_shape = np.array(range(h))
    fil_shape[fil_shape%2==0] += 1
    fil_shape = abs(fil_shape-h/2)+2
    
    p = int(fil_shape[0]/2)
    
    tmp_img = np.zeros([h+2*p, w+2*p])
    tmp_img[p:p+h, p:p+w] = img
    
    out = np.zeros_like(img)
    
    for i in range(h):
        
        for (j, f) in zip(range(w), fil_shape):
            
            filt = gaussian_filter(int(f), f/fil_shape[0])
            f_h, f_w = filt.shape
            mid_point = int(f_h/2)
            out[i,j] = (tmp_img[int(p+i-f_h):int(p+i+f_h+1), int(p+j-f_w):int(p+j+f_w+1)]*filt).sum()
    return out
```
